```python
# Imports
import yfinance as yf
```

```python
# Define the tickers for Bitcoin, Ethereum, and Solana on yfinance
tickers = ["BTC-USD", "ETH-USD", "SOL-USD"]

# Set the time period for the data (for last 3 years)
start_date = "2021-01-01"
end_date = "2024-12-21"
```

```python
# Download the data using yfinance
crypto_data = yf.download(tickers, start=start_date, end=end_date, interval='1d')

# Display the first few rows of the data
crypto_data.head()
```

```python
# Getting a look at the data
df = crypto_data[['Adj Close','Volume']]
df.head()
```

```python
# Flattening the columns
df.columns = df.columns.map('_'.join)
```

```python
# Checking the data
df.head()
```

```python
df = df.reset_index()
```

```python
# Save the data to a CSV file for later use
df.to_csv("crypto_data.csv")
```
